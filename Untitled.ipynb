### Feature Engineering

#### Missing Value for the Train Data

```python
## I will capture the missing value
## Fisrtly i will focus on categorical features which are missing

feature_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes=='O']


## Step -2 print the feature name and the perentage of missing value

for feature in feature_nan:
    print("{}: {}% missing values".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
## Replacing missing value with a new label
def replace_cat_feature(dataset,feature_nan):
    data=dataset.copy()
    data[feature_nan]=data[feature_nan].fillna('Missing')
    return data

dataset=replace_cat_feature(dataset,feature_nan)

dataset[feature_nan].isnull().sum()
```

```python
dataset.head()
```

```python
## Finding missing value for the numerial feature
numerical_with_na = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes!='O']

## I will print the numerical nan variable and percentage of the missing value
for feature in numerical_with_na:
    print("{}: {}% missing value".format(feature,np.around(dataset[feature].isnull().mean(),4)))
```

```python
## Replacing the numerical missing value
## As dataset has many outlier, missing value can be replace by median or mode. Here i will use median to replace missing value for the numerical feature.

for feature in numerical_with_na:
    ## I will use meadian since there are outlier
    median_value=dataset[feature].median()
    
    ## create a new feature to capture nan value
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)
    
dataset[numerical_with_na].isnull().sum()
```

```python
dataset.head(50)
```

```python
dataset[feature+'nan'].unique()
```

```python
## Temporal Variable (Time Variable)

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    
    dataset[feature]=dataset['YrSold'] - dataset[feature]
```

```python
dataset[['YearBuilt','YearRemodAdd','GarageYrBlt']].head(10)
```

## Numerical Distribution

```python
num_feature = ['LotFrontage', 'LotArea', '1stFlrSF','GrLivArea','SalePrice']

for feature in num_feature:
    dataset[feature]=np.log(dataset[feature])

dataset.head()
```

### Handling Rare Categorical value
I will remove categorical variable that are present in the variable whose percent is less 1% of the observation

```python
categorical_feature = [feature for feature in dataset.columns if dataset[feature].dtypes == 'O']
```

```python
categorical_feature
```

```python
## Here i will replace categrical feature who value is less than 1%
for feature in categorical_feature:
    temp = dataset.groupby(feature)["SalePrice"].count()/len(dataset)
    temp_def=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_def),dataset[feature],"Rare_Var")
```

```python
dataset.head(100)
```

```python
for feature in categorical_feature:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```

### Feature Scaling

```python
feature_scaler = [feature for feature in dataset.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(dataset[feature_scaler])
```

```python
## Tranform the dataset and add on ID and SalePrice
data=pd.concat([dataset[['Id','SalePrice']].reset_index(drop=True),
                       pd.DataFrame(scaler.transform(dataset[feature_scaler]),columns=feature_scaler)],
               axis=1)
```

```python
data.to_csv('x_train.csv',index=False)
```

### Feature Selection

```python
## Import Important feature selection and machine learning algorithm Library
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
train_data = pd.read_csv('x_train.csv')
```

```python
train_data.head()
```

```python
## I will capture the dependend feature

y_train = train_data[['SalePrice']]
```

```python
## I will drop dependend feature from the dataset

x_train = train_data.drop(['Id','SalePrice'],axis = 1)
```

```python
### Apply Feature selection 
# Firstly, i specify the lasso regression model,
# Select a suitable alpha 
# the bigger the alpha value the less features that will be selected

# Secondly, I use the SelectFromModel object from sklearn,
# it will select the feature which coefficient are non zero

feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0)) # remember to set the seed, the random state in this function
feature_sel_model.fit(x_train, y_train)
```

```python
## .get_support will show us that which independent feature is important for the dependend feature.
## Here true tells that this feature is important.

feature_sel_model.get_support()
```

```python
# let's print the number of total and selected features

# this is how we can make a list of the selected features
selected_feat = x_train.columns[(feature_sel_model.get_support())]

# let's print some stats
print('total features: {}'.format((x_train.shape[1])))
print('selected features: {}'.format(len(selected_feat)))
print('features with coefficients shrank to zero: {}'.format(np.sum(feature_sel_model.estimator_.coef_== 0)))
```
